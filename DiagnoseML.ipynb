```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt    
import warnings
```

```python
warnings.filterwarnings("ignore")
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
df = pd.read_csv("Disease_symp.csv")
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.drop(columns=['Symptom_11','Symptom_12','Symptom_13','Symptom_14','Symptom_15','Symptom_16','Symptom_17'], inplace=True)
df
```

```python
df.fillna('no_symptom', inplace=True)
df.head(8)
```

```python
symptom = [f'Symptom_{i}' for i in range(1, 11)]
df['Symptoms'] = df[symptom].apply(lambda row: ','.join(row.dropna()), axis=1)
df
```

```python
new_df = df[['Disease', 'Symptoms']]
new_df.head()
```

```python
new_df['Symptoms'] = new_df['Symptoms'].str.replace(' ', '').str.strip()
new_df['Symptoms'] = new_df['Symptoms'].str.replace('_', ' ').str.strip()
```

```python
new_df.isnull().sum()
```

```python
y = new_df.iloc[:,0]
y
```

```python
x = new_df.iloc[:,1]
x
```

```python
y_unique = pd.unique(y)
y_unique
```

```python
all_symptoms = new_df['Symptoms'].str.split(',')
flat = all_symptoms.explode()
x_unique = pd.unique(flat)
x_unique
```

```python
print("Diseases : ",len(y_unique))
print("Symptoms : ",len(x_unique))
```

```python
from sklearn.preprocessing import LabelEncoder
le_disease = LabelEncoder()
new_df['Disease'] = le_disease.fit_transform(y)
new_df.head()
```

```python
label_mapping = dict(zip(le_disease.classes_,le_disease.transform(le_disease.classes_)))
if label_mapping:
    print("Label Encoding for Diseases:")
    for disease, encoded_value in label_mapping.items():
        print(f"{disease}: {encoded_value}")
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
new_df['Symptoms'] = new_df['Symptoms'].str.split(',')
mlb = MultiLabelBinarizer()
symptom_encoded = mlb.fit_transform(new_df['Symptoms'])
symptom_df = pd.DataFrame(symptom_encoded, columns=mlb.classes_)
final_df = pd.concat([new_df['Disease'], symptom_df], axis=1)
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
x_encoded = final_df.drop(columns=['Disease']) 
y_encoded = final_df['Disease']
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.3, random_state=42,stratify=y_encoded)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV

model = SVC()
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(x_train, y_train)
print(f"Best parameters: {grid_search.best_params_}")
```

```python
model = make_pipeline(StandardScaler(), SVC(kernel='linear', C=0.1,random_state=42))
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_svm = accuracy_score(y_test, y_pred)
print("SVM Accuracy:", accuracy_svm)
print("Classification Report for SVM:")
print(classification_report(y_test, y_pred,zero_division=1))
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
if train_accuracy > test_accuracy:
    print("The model may be overfitting.")
elif train_accuracy < test_accuracy:
    print("The model may be underfitting.")
else:
    print("The model is performing consistently on both training and test sets.")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Classification Report:")
print(classification_report(y_test, y_test_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(model, x_encoded, y_encoded, cv=5)
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())
```

GIVING SOME INPUT SYMPTOMS

```python
input_symptoms = ['weight loss','restlessness','lethargy','irregular sugar level','blurred and distorted vision','obesity','excessive hunger','increased appetite','polyuria']
input_encoded = mlb.transform([input_symptoms])
prediction = model.predict(input_encoded)
predicted_disease = le_disease.inverse_transform(prediction)
print(f"The predicted disease is: {predicted_disease[0]}")
```

```python
predicted_disease_str = predicted_disease[0].strip()
descriptions_df = pd.read_csv('Description.csv')
precautions_df = pd.read_csv('Precaution.csv')
descriptions_df['Disease'] = descriptions_df['Disease'].str.strip()
precautions_df['Disease'] = precautions_df['Disease'].str.strip()
matching_description = descriptions_df.loc[descriptions_df['Disease'] == predicted_disease_str, 'Description']

if len(matching_description) > 0:
    description = matching_description.values[0]
else:
    description = "Description not found"
matching_precautions = precautions_df.loc[precautions_df['Disease'] == predicted_disease_str, ['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
if len(matching_precautions) > 0:
    precautions = matching_precautions.values[0]
else:
    precautions = ["Precautions not available"]
print(f"Disease: {predicted_disease_str}\n")
print(f"Description: {description}\n")
print("Precautions:\n " + '\n '.join(precautions))
```
